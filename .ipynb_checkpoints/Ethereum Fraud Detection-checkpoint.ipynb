# Ethereum Fraud Detection Using Machine Learning

In this notebook, Ethereum fraud is detected using Data Science and Machine Learning Technqiues. The data is first cleaned and analysed using Exploratory Data Analysis. Then three models will be used to predict fraud and their performances are compared and evaluated.

Machine Learning Models used in this project include:
- Logistic Regression
- XGBoost
- Multilayer Perceptron

Stages in this project involve:

- Exploratory Data Analysis (EDA)
- Data Preprocessing:
    - Feature Selection - Dropping zero and highly correlated columns
    - Data Train, Test Split
    - Filling missing values
    - Encoding Categorical Features
    - Rebalancing - SMOTE
    - Normalisation - StandardScaler
    - Dimensionality Reduction - PCA
- Model Implementation 
- Model Evaluation
- Results Comparision

#### Removing Highly Correlated Columns

```python
sns.heatmap(df.corr())
```

```python
#Calculate Corr Matrix

corr = df.corr().abs()

# Create and apply mask
# np.ones_like creates a matrix of the same shape as corr, filled with True values (since dtype=bool is specified).
#This function extracts the upper triangle of the matrix (including the diagonal), converting all values below the diagonal to False. 
#The result is a boolean matrix that has True for the upper triangle (and diagonal) and False elsewhere.
# .mask() method is used to hide (or "mask") values in the corr matrix where the corresponding mask is True. Specifically, it sets those values to NaN. 
#The result is a matrix where only the lower triangle (below the diagonal) retains the original correlation values; 
#the upper triangle (including the diagonal) is replaced with NaN.


mask = np.triu(np.ones_like(corr, dtype=bool))
tri_df = corr.mask(mask)

tri_df
```

```python
threshold = 0.8

#Find Values that meet threshold 

#The any() function checks if any value in the boolean Series is True. 
#If there is at least one True value (i.e., one correlation value in column c that is greater than threshold), 
# this condition is satisfied.

to_drop = [c for c in tri_df.columns if any (tri_df[c] > threshold)]

to_drop
```

```python
# Drop Highly Correlated Features

df = df.drop(to_drop, axis=1)
```

### Encoding Categorical Data

ERC20_most_sent_token_type and ERC20_most_rec_token_type are categorical which means they need to be encoded to be able to be fed into the models. Here I am using One Hot Encoding to encode values and make sure they are not ordinal.

```python
train_X = df_encoded.drop(columns='FLAG')
train_y = df_encoded['FLAG']
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_X, train_y, test_size = 0.3, random_state = 0)
```

### Rebalancing

```python
df_encoded['FLAG'].value_counts()
```

#### SMOTE (Synthetic Minority Oversampling Technique)

```python
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

```python
print(X_train_sm.shape)
print(y_train_sm.shape)
```

```python
y_train_sm.value_counts()
```

### Normalisation 

The data in this dataset has a wide range and each column having a different ranges. Given the inconsistency, the data is standardized

```python
scaler = StandardScaler()

std_X_train = scaler.fit_transform(X_train_sm)
std_X_test = scaler.fit_transform(X_test)
```

```python
print(std_X_train)
print(std_X_test)
```

### Dimensionality Reduction Using Principle Component Analysis (PCA)

Searching for the optimal number of components using the total explainedf variance

```python
pca = PCA()
X_train_pca = pca.fit_transform(std_X_train)

total_explained_variance = pca.explained_variance_ratio_.cumsum()
pca_optimal_n = (total_explained_variance >= 0.90).argmax() + 1

pca_optimal_n
```

```python
pca_opt = PCA(n_components = pca_optimal_n)

X_train_pca = pca_opt.fit_transform(std_X_train)
X_test_pca = pca_opt.fit_transform(std_X_test)
```

```python
print(X_train_pca)
```

### Multilayer Perceptron

Using Tensorflow and Kera to nuild a simple feedforward neural network (multilayer perceptron)

```python
#Initialize the model
model = Sequential()

# first dense layer with 64 units and ReLU activation
model.add(Dense(64, activation='relu', input_shape=(X_train_pca.shape[1],)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))

# Add the output layer with 1 unit and sigmoid activation (for binary classification)
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train_pca, y_train_sm, epochs=100, batch_size=32, validation_split=0.2)

# Evaluate the Model
test_loss, test_accuracy = model.evaluate(X_train_pca, y_train_sm)
print(f"Test Accuracy: {test_accuracy:.4f}")
```

```python
history_df = pd.DataFrame(history.history)
history_df.loc[:, ['loss', 'val_loss']].plot(title="Cross-entropy")
```

Hyperparameter Tuning

```python
def create_model(learning_rate=0.01, dropout_rate=0.0, activation='relu', neurons=32):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, input_shape=(X_train_pca.shape[1],)))
    model.add(Dropout(dropout_rate))
    model.add(Dense(neurons, activation=activation))
    model.add(Dropout(dropout_rate))
    model.add(Dense(neurons, activation=activation))
    model.add(Dense(1, activation='sigmoid'))
    
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
#Wrapping model using KerasClassifier
model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
param_grid = {
    'batch_size': [16, 32, 64],
    'epochs': [50, 100, 125],
    'learning_rate': [0.001, 0.01, 0.1],
    'dropout_rate': [0.0, 0.2, 0.5],
    'neurons': [32, 64, 128],
    'activation': ['relu', 'tanh']
}
```

```python
random_search = RandomizedSearchCV(estimator=model, param_distributions=param_grid, scoring='accuracy', cv=5, verbose=1)
best_param = random_search.fit(X_train_pca, y_train_sm)

print("Best parameters found: ", best_param.best_params_)
print("Best accuracy: ", best_param.best_score_)
```

```python
best_model = create_model(learning_rate=best_param.best_params_['learning_rate'],
                          dropout_rate=best_param.best_params_['dropout_rate'],
                          activation=best_param.best_params_['activation'],
                          neurons=best_param.best_params_['neurons'])

history = best_model.fit(X_train_pca, y_train_sm,
                         batch_size=best_param.best_params_['batch_size'],
                         epochs=best_param.best_params_['epochs'],
                         )
```

```python
MP_HP_y_pred = best_model.predict(X_test_pca)
```

```python
MP_HP_y_pred_binary = list(map(lambda x: 0 if x<0.5 else 1, MP_HP_y_pred))
```
